## 2. Load EEG Data

> **Replace** the synthetic data section below with your real data loader.  
> Supported formats via MNE: `.edf`, `.bdf`, `.fif`, `.set` (EEGLAB), `.cnt`, `.vhdr` (BrainVision).

**Expected data structure:**
- `n_subjects` recordings, each with shape `(n_channels, n_timepoints)`
- Labels: `0 = Control`, `1 = Alzheimer's`, `2 = FTD`

```python
# ─────────────────────────────────────────────────────────────────────
# OPTION A: Load your real EEG files (uncomment and adapt)
# ─────────────────────────────────────────────────────────────────────
# import glob, os
# 
# def load_eeg_files(data_dir, label_map):
#     """Load all EDF files from directory structure.
#     label_map example: {'control': 0, 'alzheimer': 1, 'ftd': 2}
#     Expects: data_dir/group_name/*.edf
#     """
#     raws, labels = [], []
#     for group_name, label in label_map.items():
#         pattern = os.path.join(data_dir, group_name, '*.edf')
#         for fpath in sorted(glob.glob(pattern)):
#             raw = mne.io.read_raw_edf(fpath, preload=True, verbose=False)
#             raws.append(raw)
#             labels.append(label)
#             print(f'  Loaded: {os.path.basename(fpath)} -> label {label}')
#     return raws, np.array(labels)
#
# DATA_DIR = '/path/to/your/eeg/data'
# LABEL_MAP = {'control': 0, 'alzheimer': 1, 'ftd': 2}
# raw_list, true_labels = load_eeg_files(DATA_DIR, LABEL_MAP)

# ─────────────────────────────────────────────────────────────────────
# OPTION B: Synthetic data (for demonstration / development)
# ─────────────────────────────────────────────────────────────────────
np.random.seed(42)

SFREQ       = 256        # Hz
N_CHANNELS  = 19         # Standard 10-20 system
DURATION    = 60         # seconds per recording
N_PER_GROUP = 15         # subjects per group
N_SUBJECTS  = N_PER_GROUP * 3

CH_NAMES = [
    'Fp1','Fp2','F3','F4','F7','F8','Fz',
    'C3','C4','Cz','P3','P4','P7','P8','Pz',
    'O1','O2','T3','T4'
]

def make_eeg_signal(sfreq, duration, n_channels, group):
    """Generate realistic synthetic EEG with group-specific spectral properties."""
    t = np.linspace(0, duration, int(sfreq * duration))
    n_t = len(t)
    data = np.zeros((n_channels, n_t))

    # Group-specific band power profiles (microvolts)
    profiles = {
        #           delta  theta  alpha  beta   gamma
        'control':  [2.0,  1.5,   4.0,   2.5,   1.0],
        'alzheimer':[4.5,  3.5,   1.5,   1.5,   0.5],  # increased slow, reduced alpha
        'ftd':      [3.5,  4.0,   2.0,   3.5,   1.5],  # frontal theta dominance
    }
    freqs = [(1,4), (4,8), (8,13), (13,30), (30,45)]
    amps  = profiles[group]

    for ch in range(n_channels):
        sig = np.zeros(n_t)
        for (f_lo, f_hi), amp in zip(freqs, amps):
            f_center = np.random.uniform(f_lo, f_hi)
            sig += amp * np.sin(2 * np.pi * f_center * t + np.random.uniform(0, 2*np.pi))
        sig += np.random.normal(0, 0.8, n_t)  # background noise
        # Simulate occasional muscle artifact
        if np.random.rand() < 0.3:
            art_start = np.random.randint(0, n_t - sfreq)
            sig[art_start:art_start + sfreq//4] += np.random.normal(0, 15, sfreq//4)
        data[ch] = sig

    return data * 1e-6  # convert to Volts for MNE

# Create MNE Raw objects
info = mne.create_info(ch_names=CH_NAMES, sfreq=SFREQ, ch_types='eeg')
info.set_montage('standard_1020')

raw_list, true_labels = [], []
group_names = ['control'] * N_PER_GROUP + ['alzheimer'] * N_PER_GROUP + ['ftd'] * N_PER_GROUP
label_map   = {'control': 0, 'alzheimer': 1, 'ftd': 2}

for grp in group_names:
    data = make_eeg_signal(SFREQ, DURATION, N_CHANNELS, grp)
    raw  = mne.io.RawArray(data, info, verbose=False)
    raw_list.append(raw)
    true_labels.append(label_map[grp])

true_labels = np.array(true_labels)
print(f'Loaded {N_SUBJECTS} subjects: {N_PER_GROUP} Control / {N_PER_GROUP} Alzheimer / {N_PER_GROUP} FTD')
print(f'Recording: {N_CHANNELS} channels, {DURATION}s @ {SFREQ} Hz')
```

## 3. Artifact Removal Pipeline

Three-stage cleaning:
1. **Bandpass filter** – remove DC drift and high-frequency noise  
2. **Notch filter** – remove powerline interference (50 Hz Europe / 60 Hz USA)  
3. **ICA** – remove ocular (EOG) and cardiac (ECG) artifacts  
4. **Epoch rejection** – drop epochs exceeding amplitude threshold

```python
def remove_artifacts(raw, 
                     l_freq=0.5, h_freq=45.0,
                     notch_freq=50.0,
                     n_ica_components=15,
                     epoch_duration=4.0,
                     reject_threshold=150e-6):
    """
    Full artifact removal pipeline.
    
    Parameters
    ----------
    raw              : mne.io.Raw
    l_freq           : High-pass cutoff (Hz). 0.5 for EEG, 1.0 if drift is severe.
    h_freq           : Low-pass cutoff (Hz). 45 keeps gamma, reduces EMG.
    notch_freq       : Powerline frequency. 50 Hz (EU/Asia), 60 Hz (USA).
    n_ica_components : ICA components to fit (rule of thumb: ~n_channels * 0.8).
    epoch_duration   : Length of epochs for rejection (seconds).
    reject_threshold : Peak-to-peak amplitude limit (Volts). 150 µV is standard.
    
    Returns
    -------
    clean_data : np.ndarray, shape (n_channels, n_timepoints_clean)
    """
    raw = raw.copy()
    
    # ── Step 1: Bandpass filter ──────────────────────────────────────
    raw.filter(l_freq=l_freq, h_freq=h_freq, 
               method='fir', fir_window='hamming',
               verbose=False)
    
    # ── Step 2: Notch filter (powerline) ────────────────────────────
    raw.notch_filter(freqs=notch_freq, verbose=False)
    
    # ── Step 3: ICA artifact removal ─────────────────────────────────
    # ICA requires high-pass >= 1 Hz for stability; use a filtered copy
    raw_for_ica = raw.copy().filter(l_freq=1.0, h_freq=None, verbose=False)
    
    ica = ICA(
        n_components=min(n_ica_components, len(raw.ch_names) - 1),
        method='fastica',
        random_state=42,
        max_iter=400
    )
    ica.fit(raw_for_ica, verbose=False)
    
    # Auto-detect EOG-like components via frontal channel correlation
    frontal_chs = [ch for ch in ['Fp1', 'Fp2'] if ch in raw.ch_names]
    eog_indices = []
    if frontal_chs:
        try:
            eog_idx, eog_scores = ica.find_bads_eog(
                raw, ch_name=frontal_chs[0], threshold=2.5, verbose=False
            )
            eog_indices.extend(eog_idx)
        except Exception:
            pass  # Skip if EOG detection fails
    
    # Fallback: flag components with high kurtosis (muscle) or low freq (EOG)
    sources = ica.get_sources(raw_for_ica).get_data()
    kurt = np.array([kurtosis(src) for src in sources])
    high_kurt_idx = np.where(kurt > 5.0)[0].tolist()  # muscle artifact
    
    exclude = list(set(eog_indices + high_kurt_idx))[:3]  # cap at 3 components
    ica.exclude = exclude
    ica.apply(raw, verbose=False)
    
    # ── Step 4: Epoch-based amplitude rejection ──────────────────────
    events = mne.make_fixed_length_events(
        raw, duration=epoch_duration, overlap=0.0
    )
    epochs = mne.Epochs(
        raw, events,
        tmin=0, tmax=epoch_duration,
        baseline=None,
        reject={'eeg': reject_threshold},
        flat={'eeg': 1e-8},
        verbose=False,
        preload=True
    )
    
    if len(epochs) == 0:
        print('  WARNING: All epochs rejected! Relaxing threshold to 300 µV.')
        epochs = mne.Epochs(
            raw, events,
            tmin=0, tmax=epoch_duration,
            baseline=None,
            reject={'eeg': 300e-6},
            verbose=False, preload=True
        )
    
    # Reconstruct continuous signal from clean epochs
    clean_data = epochs.get_data().transpose(1, 0, 2).reshape(len(raw.ch_names), -1)
    return clean_data


# ── Apply to all subjects ────────────────────────────────────────────
print('Cleaning EEG recordings...')
clean_data_list = []
for i, raw in enumerate(raw_list):
    group = list(label_map.keys())[true_labels[i]]
    clean = remove_artifacts(raw)
    clean_data_list.append(clean)
    if (i + 1) % 5 == 0:
        print(f'  Processed {i+1}/{N_SUBJECTS} subjects...')

print(f'\nCleaning complete. Output shape per subject: {clean_data_list[0].shape}')
```

## 4. Feature Extraction

### 4a. Statistical Features (time-domain)

```python
def extract_statistical_features(data):
    """
    Extract time-domain statistical features per channel.
    
    Features per channel:
    - Mean, Variance, Std, RMS
    - Skewness, Kurtosis  (shape of amplitude distribution)
    - Peak-to-peak, Zero-crossing rate
    - Hjorth parameters (Activity, Mobility, Complexity)
    
    Returns: 1D feature vector of length n_channels * 11
    """
    n_channels, n_times = data.shape
    feats = []
    
    for ch in range(n_channels):
        x = data[ch]
        dx  = np.diff(x)
        ddx = np.diff(dx)
        
        # Basic statistics
        mean_val = np.mean(x)
        var_val  = np.var(x)
        std_val  = np.std(x)
        rms      = np.sqrt(np.mean(x**2))
        skew_val = skew(x)
        kurt_val = kurtosis(x)
        ptp      = np.ptp(x)  # peak-to-peak
        
        # Zero-crossing rate
        zcr = np.sum(np.diff(np.sign(x)) != 0) / len(x)
        
        # Hjorth parameters
        activity   = var_val
        mobility   = np.sqrt(np.var(dx) / (var_val + 1e-10))
        complexity = np.sqrt(np.var(ddx) / (np.var(dx) + 1e-10)) / (mobility + 1e-10)
        
        feats.extend([
            mean_val, var_val, std_val, rms,
            skew_val, kurt_val, ptp, zcr,
            activity, mobility, complexity
        ])
    
    return np.array(feats)


print('Statistical features per subject:')
test_feats = extract_statistical_features(clean_data_list[0])
print(f'  Feature vector length: {len(test_feats)} ({N_CHANNELS} channels × 11 features)')
```

### 4b. Frequency-Domain Features (PSD & Band Powers)

```python
# EEG frequency bands
BANDS = {
    'delta': (0.5, 4),
    'theta': (4,   8),
    'alpha': (8,  13),
    'beta':  (13, 30),
    'gamma': (30, 45),
}

def extract_frequency_features(data, sfreq):
    """
    Extract frequency-domain features using Welch's PSD method.
    
    Features per channel:
    - Absolute band power: delta, theta, alpha, beta, gamma
    - Relative band power: each band / total power
    - Clinically important ratios:
        * Theta/Alpha  (elevated in AD)
        * (Delta+Theta)/(Alpha+Beta)  (slow-wave ratio, elevated in AD)
        * Alpha/Beta
    - Spectral edge frequency (95% of power)
    - Median frequency
    - Spectral entropy
    
    Returns: 1D feature vector of length n_channels * (10 + 5 + 3 + 3)
    """
    n_channels = data.shape[0]
    feats = []
    
    nperseg = min(int(sfreq * 4), data.shape[1])  # 4-second windows
    
    for ch in range(n_channels):
        freqs, psd = signal.welch(
            data[ch], fs=sfreq,
            nperseg=nperseg, noverlap=nperseg//2,
            window='hann'
        )
        
        # Absolute band powers
        band_powers = {}
        for band, (lo, hi) in BANDS.items():
            idx = np.logical_and(freqs >= lo, freqs <= hi)
            band_powers[band] = np.trapz(psd[idx], freqs[idx])
        
        total_power = sum(band_powers.values()) + 1e-10
        abs_powers  = list(band_powers.values())
        
        # Relative band powers
        rel_powers  = [p / total_power for p in abs_powers]
        
        # Clinical ratios
        ta_ratio    = band_powers['theta'] / (band_powers['alpha'] + 1e-10)
        slow_ratio  = (band_powers['delta'] + band_powers['theta']) / \
                      (band_powers['alpha'] + band_powers['beta'] + 1e-10)
        ab_ratio    = band_powers['alpha'] / (band_powers['beta'] + 1e-10)
        
        # Spectral edge frequency (SEF95)
        cumulative = np.cumsum(psd)
        sef95_idx  = np.searchsorted(cumulative, 0.95 * cumulative[-1])
        sef95      = freqs[min(sef95_idx, len(freqs) - 1)]
        
        # Median frequency
        med_idx    = np.searchsorted(cumulative, 0.5 * cumulative[-1])
        med_freq   = freqs[min(med_idx, len(freqs) - 1)]
        
        # Spectral entropy (Shannon entropy of normalized PSD)
        psd_norm   = psd / (psd.sum() + 1e-10)
        spec_entr  = -np.sum(psd_norm * np.log2(psd_norm + 1e-10))
        
        feats.extend(
            abs_powers + rel_powers +
            [ta_ratio, slow_ratio, ab_ratio,
             sef95, med_freq, spec_entr]
        )
    
    return np.array(feats)


test_freq_feats = extract_frequency_features(clean_data_list[0], SFREQ)
n_freq_per_ch   = 5 + 5 + 3 + 3  # abs + rel + ratios + spectral
print(f'Frequency features per subject: {len(test_freq_feats)} ({N_CHANNELS} channels × {n_freq_per_ch} features)')
```

### 4c. Build Full Feature Matrix

```python
print('Extracting all features...')
all_features = []

for i, data in enumerate(clean_data_list):
    stat_feats = extract_statistical_features(data)
    freq_feats = extract_frequency_features(data, SFREQ)
    combined   = np.concatenate([stat_feats, freq_feats])
    all_features.append(combined)

X = np.array(all_features)

# Handle NaN/Inf
X = np.nan_to_num(X, nan=0.0, posinf=0.0, neginf=0.0)

# Feature names for interpretability
stat_feat_names = [f'ch{c}_{f}' 
                   for c in range(N_CHANNELS)
                   for f in ['mean','var','std','rms','skew','kurt','ptp','zcr','activity','mobility','complexity']]

freq_feat_names = [f'ch{c}_{band}_{t}'
                   for c in range(N_CHANNELS)
                   for band in list(BANDS.keys()) + [f'rel_{b}' for b in BANDS.keys()]
                   for t in [''] 
                  ] + [f'ch{c}_{r}' 
                       for c in range(N_CHANNELS)
                       for r in ['ta_ratio','slow_ratio','ab_ratio','sef95','med_freq','spec_entr']]

print(f'\nFeature matrix shape: {X.shape}')
print(f'  - {len(stat_feat_names)} statistical features')
print(f'  - {X.shape[1] - len(stat_feat_names)} frequency features')
print(f'  - {X.shape[0]} subjects')
```

---
## 5. Preprocessing & Dimensionality Reduction

```python
# ── Standardize features ─────────────────────────────────────────────
scaler   = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ── PCA: reduce to 95% explained variance ────────────────────────────
pca_full     = PCA(random_state=42)
X_pca_full   = pca_full.fit_transform(X_scaled)
cumvar        = np.cumsum(pca_full.explained_variance_ratio_)
n_components  = np.argmax(cumvar >= 0.95) + 1

pca      = PCA(n_components=n_components, random_state=42)
X_pca    = pca.fit_transform(X_scaled)
X_pca_2d = X_pca_full[:, :2]  # for visualization

print(f'PCA: {n_components} components explain 95% of variance')
print(f'Feature space: {X.shape[1]}D → {n_components}D')

# Plot explained variance
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].bar(range(1, 21), pca_full.explained_variance_ratio_[:20] * 100,
            color='steelblue', edgecolor='white')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Explained Variance (%)')
axes[0].set_title('Scree Plot (top 20 PCs)')

axes[1].plot(range(1, len(cumvar)+1), cumvar * 100, 'o-', color='steelblue', ms=3)
axes[1].axhline(95, color='red', linestyle='--', label='95% threshold')
axes[1].axvline(n_components, color='orange', linestyle='--', label=f'n={n_components}')
axes[1].set_xlabel('Number of Components')
axes[1].set_ylabel('Cumulative Explained Variance (%)')
axes[1].set_title('Cumulative Explained Variance')
axes[1].legend()

plt.tight_layout()
plt.show()
```

---
## 6. K-Means Clustering

### 6a. Optimal k via Elbow + Silhouette

```python
k_range      = range(2, 9)
inertias     = []
silhouettes  = []
ch_scores    = []
db_scores    = []

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=20, max_iter=500)
    labels = km.fit_predict(X_pca)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_pca, labels))
    ch_scores.append(calinski_harabasz_score(X_pca, labels))
    db_scores.append(davies_bouldin_score(X_pca, labels))

fig, axes = plt.subplots(2, 2, figsize=(13, 8))
fig.suptitle('K-Means: Optimal k Selection', fontsize=14, fontweight='bold')

axes[0,0].plot(k_range, inertias, 'o-', color='steelblue')
axes[0,0].set_title('Elbow Method (Inertia)')
axes[0,0].set_xlabel('k'); axes[0,0].set_ylabel('Inertia')

axes[0,1].plot(k_range, silhouettes, 's-', color='seagreen')
axes[0,1].axvline(k_range[np.argmax(silhouettes)], color='red', linestyle='--')
axes[0,1].set_title('Silhouette Score (↑ better)')
axes[0,1].set_xlabel('k'); axes[0,1].set_ylabel('Score')

axes[1,0].plot(k_range, ch_scores, '^-', color='darkorange')
axes[1,0].axvline(k_range[np.argmax(ch_scores)], color='red', linestyle='--')
axes[1,0].set_title('Calinski-Harabasz (↑ better)')
axes[1,0].set_xlabel('k'); axes[1,0].set_ylabel('Score')

axes[1,1].plot(k_range, db_scores, 'd-', color='orchid')
axes[1,1].axvline(k_range[np.argmin(db_scores)], color='red', linestyle='--')
axes[1,1].set_title('Davies-Bouldin (↓ better)')
axes[1,1].set_xlabel('k'); axes[1,1].set_ylabel('Score')

plt.tight_layout()
plt.show()

best_k = 3  # We know we have 3 groups; use metrics above to verify
print(f'\nBest silhouette at k={k_range[np.argmax(silhouettes)]}')
print(f'Using k={best_k} (matching number of clinical groups)')
```

### 6b. Fit Final K-Means & Evaluate

```python
# Fit K-Means with multiple restarts
kmeans = KMeans(
    n_clusters=best_k,
    random_state=42,
    n_init=50,        # run 50 times, keep best
    max_iter=1000,
    algorithm='lloyd'
)
km_labels = kmeans.fit_predict(X_pca)

# Metrics
km_sil = silhouette_score(X_pca, km_labels)
km_ari = adjusted_rand_score(true_labels, km_labels)
km_ch  = calinski_harabasz_score(X_pca, km_labels)
km_db  = davies_bouldin_score(X_pca, km_labels)

print('K-Means Results')
print('=' * 40)
print(f'Silhouette Score:       {km_sil:.4f}  (range -1→1, higher=better)')
print(f'Adjusted Rand Index:    {km_ari:.4f}  (0=random, 1=perfect)')
print(f'Calinski-Harabasz:      {km_ch:.2f}')
print(f'Davies-Bouldin:         {km_db:.4f}  (lower=better)')
print()
print('Cluster composition:')
group_names_list = ['Control', 'Alzheimer', 'FTD']
comp_df = pd.DataFrame({'True': true_labels, 'KMeans': km_labels})
for cluster in range(best_k):
    subset = comp_df[comp_df['KMeans'] == cluster]['True']
    counts = {group_names_list[g]: (subset == g).sum() for g in range(3)}
    dominant = max(counts, key=counts.get)
    print(f'  Cluster {cluster}: {dict(counts)}  → likely {dominant}')
```

---
## 7. Gaussian Mixture Model (GMM) Clustering

### 7a. Select Best Covariance Type via BIC/AIC

```python
cov_types   = ['full', 'tied', 'diag', 'spherical']
bic_results = {ct: [] for ct in cov_types}
aic_results = {ct: [] for ct in cov_types}

for ct in cov_types:
    for k in k_range:
        gmm = GaussianMixture(n_components=k, covariance_type=ct,
                              random_state=42, n_init=5, max_iter=300)
        gmm.fit(X_pca)
        bic_results[ct].append(gmm.bic(X_pca))
        aic_results[ct].append(gmm.aic(X_pca))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
colors = ['steelblue', 'seagreen', 'darkorange', 'orchid']

for (ct, color) in zip(cov_types, colors):
    axes[0].plot(k_range, bic_results[ct], 'o-', label=ct, color=color)
    axes[1].plot(k_range, aic_results[ct], 'o-', label=ct, color=color)

for ax, title in zip(axes, ['BIC (↓ better)', 'AIC (↓ better)']):
    ax.set_title(f'GMM {title}')
    ax.set_xlabel('n_components')
    ax.set_ylabel('Score')
    ax.legend()

plt.suptitle('GMM Model Selection: BIC & AIC', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

# Find best configuration
best_bic = min((min(v), ct) for ct, v in bic_results.items())
best_cov = best_bic[1]
best_k_gmm = k_range[np.argmin(bic_results[best_cov])]
print(f'Best GMM: covariance_type="{best_cov}", n_components={best_k_gmm} (via BIC)')
print(f'We will use n_components={best_k} to match the 3 clinical groups')
```

### 7b. Fit Final GMM & Evaluate

```python
gmm = GaussianMixture(
    n_components=best_k,
    covariance_type=best_cov,
    random_state=42,
    n_init=20,
    max_iter=500,
    tol=1e-5
)
gmm.fit(X_pca)
gmm_labels = gmm.predict(X_pca)
gmm_probs  = gmm.predict_proba(X_pca)  # soft assignments

# Metrics
gmm_sil = silhouette_score(X_pca, gmm_labels)
gmm_ari = adjusted_rand_score(true_labels, gmm_labels)
gmm_ch  = calinski_harabasz_score(X_pca, gmm_labels)
gmm_db  = davies_bouldin_score(X_pca, gmm_labels)

print('GMM Results')
print('=' * 40)
print(f'Silhouette Score:       {gmm_sil:.4f}')
print(f'Adjusted Rand Index:    {gmm_ari:.4f}')
print(f'Calinski-Harabasz:      {gmm_ch:.2f}')
print(f'Davies-Bouldin:         {gmm_db:.4f}')
print(f'Log-Likelihood:         {gmm.score(X_pca):.4f}')
print()
print('Cluster composition:')
comp_df_gmm = pd.DataFrame({'True': true_labels, 'GMM': gmm_labels})
for cluster in range(best_k):
    subset = comp_df_gmm[comp_df_gmm['GMM'] == cluster]['True']
    counts = {group_names_list[g]: (subset == g).sum() for g in range(3)}
    dominant = max(counts, key=counts.get)
    print(f'  Cluster {cluster}: {dict(counts)}  → likely {dominant}')
```

## 8. Visualization

```python
COLORS_TRUE    = {0: '#2196F3', 1: '#F44336', 2: '#4CAF50'}
COLORS_CLUSTER = {0: '#FF9800', 1: '#9C27B0', 2: '#00BCD4'}
MARKERS_TRUE   = {0: 'o', 1: 's', 2: '^'}

fig = plt.figure(figsize=(18, 14))
gs  = gridspec.GridSpec(2, 3, figure=fig, hspace=0.4, wspace=0.35)

# ── Row 1: PCA scatter plots ─────────────────────────────────────────
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[0, 2])

# True labels
for grp, name in enumerate(group_names_list):
    idx = true_labels == grp
    ax1.scatter(X_pca_2d[idx, 0], X_pca_2d[idx, 1],
                c=COLORS_TRUE[grp], marker=MARKERS_TRUE[grp],
                s=80, label=name, alpha=0.85, edgecolors='white', lw=0.5)
ax1.set_title('True Labels (Ground Truth)', fontweight='bold')
ax1.set_xlabel(f'PC1 ({pca_full.explained_variance_ratio_[0]*100:.1f}%)')
ax1.set_ylabel(f'PC2 ({pca_full.explained_variance_ratio_[1]*100:.1f}%)')
ax1.legend(fontsize=8)

# K-Means results
for c in range(best_k):
    idx = km_labels == c
    ax2.scatter(X_pca_2d[idx, 0], X_pca_2d[idx, 1],
                c=COLORS_CLUSTER[c], s=80, label=f'Cluster {c}',
                alpha=0.85, edgecolors='white', lw=0.5)
ax2.scatter(pca_full.transform(scaler.transform(scaler.inverse_transform(pca.inverse_transform(kmeans.cluster_centers_))))[:, 0],
            pca_full.transform(scaler.transform(scaler.inverse_transform(pca.inverse_transform(kmeans.cluster_centers_))))[:, 1],
            c='black', marker='X', s=200, zorder=5, label='Centroids')
ax2.set_title(f'K-Means (ARI={km_ari:.3f}, Sil={km_sil:.3f})', fontweight='bold')
ax2.set_xlabel(f'PC1'); ax2.set_ylabel(f'PC2')
ax2.legend(fontsize=8)

# GMM results (size = certainty)
max_prob = gmm_probs.max(axis=1)
for c in range(best_k):
    idx = gmm_labels == c
    ax3.scatter(X_pca_2d[idx, 0], X_pca_2d[idx, 1],
                c=COLORS_CLUSTER[c], s=max_prob[idx] * 150,
                alpha=0.85, label=f'Cluster {c}', edgecolors='white', lw=0.5)
ax3.set_title(f'GMM (ARI={gmm_ari:.3f}, Sil={gmm_sil:.3f})\n(marker size = assignment confidence)',
              fontweight='bold', fontsize=9)
ax3.set_xlabel('PC1'); ax3.set_ylabel('PC2')
ax3.legend(fontsize=8)

# ── Row 2: Confusion matrices & comparison ───────────────────────────
ax4 = fig.add_subplot(gs[1, 0])
ax5 = fig.add_subplot(gs[1, 1])
ax6 = fig.add_subplot(gs[1, 2])

# Helper: best label matching for unsupervised
from itertools import permutations

def best_mapping(pred, true, n_clusters):
    best_acc, best_perm = -1, None
    for perm in permutations(range(n_clusters)):
        mapped = np.array([perm[p] for p in pred])
        acc    = (mapped == true).mean()
        if acc > best_acc:
            best_acc, best_perm = acc, perm
    mapped = np.array([best_perm[p] for p in pred])
    return mapped

km_mapped  = best_mapping(km_labels, true_labels, best_k)
gmm_mapped = best_mapping(gmm_labels, true_labels, best_k)

for ax, mapped, title in [
    (ax4, km_mapped,  'K-Means Confusion Matrix'),
    (ax5, gmm_mapped, 'GMM Confusion Matrix')
]:
    cm = confusion_matrix(true_labels, mapped)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                xticklabels=group_names_list,
                yticklabels=group_names_list)
    ax.set_title(title, fontweight='bold')
    ax.set_xlabel('Predicted'); ax.set_ylabel('True')

# Metric comparison bar chart
metrics     = ['Silhouette', 'ARI', 'Calinski-\nHarabasz\n(norm)', 'Davies-\nBouldin\n(inv)']
ch_max      = max(km_ch, gmm_ch)
db_max      = max(km_db, gmm_db)
km_vals     = [km_sil, km_ari, km_ch/ch_max, 1 - km_db/db_max]
gmm_vals    = [gmm_sil, gmm_ari, gmm_ch/ch_max, 1 - gmm_db/db_max]

x   = np.arange(len(metrics))
w   = 0.35
ax6.bar(x - w/2, km_vals,  w, label='K-Means', color='steelblue',  alpha=0.85)
ax6.bar(x + w/2, gmm_vals, w, label='GMM',     color='darkorange', alpha=0.85)
ax6.set_xticks(x); ax6.set_xticklabels(metrics, fontsize=8)
ax6.set_ylim(0, 1.15)
ax6.set_title('K-Means vs. GMM: Metric Comparison', fontweight='bold')
ax6.legend()
ax6.axhline(0, color='gray', lw=0.5)

plt.suptitle('EEG Clustering: Alzheimer\'s vs. FTD vs. Control', 
             fontsize=15, fontweight='bold', y=1.01)
plt.savefig('eeg_clustering_results.png', dpi=150, bbox_inches='tight')
plt.show()
print('Figure saved.')
```

## 9. Band Power Profiles per Group

```python
# Compute average band power across all channels for each subject
band_names  = list(BANDS.keys())
band_matrix = np.zeros((N_SUBJECTS, len(BANDS)))

for i, data in enumerate(clean_data_list):
    for j, (band, (lo, hi)) in enumerate(BANDS.items()):
        nperseg = min(int(SFREQ * 4), data.shape[1])
        total_bp = 0
        for ch in range(data.shape[0]):
            freqs, psd = signal.welch(data[ch], fs=SFREQ, nperseg=nperseg)
            idx = np.logical_and(freqs >= lo, freqs <= hi)
            total_bp += np.trapz(psd[idx], freqs[idx])
        band_matrix[i, j] = total_bp / data.shape[0]  # average over channels

# Normalize to relative power
band_rel = band_matrix / (band_matrix.sum(axis=1, keepdims=True) + 1e-10)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, mat, title in [
    (axes[0], band_matrix * 1e12, 'Absolute Band Power (pW²/Hz)'),
    (axes[1], band_rel * 100,     'Relative Band Power (%)'),
]:
    for grp, name, color in zip(range(3), group_names_list,
                                 ['#2196F3', '#F44336', '#4CAF50']):
        vals = mat[true_labels == grp]
        mean = vals.mean(axis=0)
        std  = vals.std(axis=0)
        ax.plot(band_names, mean, 'o-', label=name, color=color, lw=2)
        ax.fill_between(band_names, mean - std, mean + std, alpha=0.15, color=color)
    ax.set_title(title, fontweight='bold')
    ax.set_xlabel('Frequency Band')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)

plt.suptitle('EEG Band Power Profiles by Clinical Group', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 10. Feature Importance (PCA Loadings)

```python
# Top features contributing to PC1 and PC2
n_top = 15
all_feat_names = [f'stat_ch{c}_{f}' 
                  for c in range(N_CHANNELS)
                  for f in ['mean','var','std','rms','skew','kurt','ptp','zcr','act','mob','comp']
                 ] + [
                  f'freq_ch{c}_{b}_{t}'
                  for c in range(N_CHANNELS)
                  for b in list(BANDS.keys()) + ['rel_delta','rel_theta','rel_alpha','rel_beta','rel_gamma']
                  for t in ['pow']
                 ] + [
                  f'freq_ch{c}_{r}'
                  for c in range(N_CHANNELS)
                  for r in ['ta_ratio','slow_ratio','ab_ratio','sef95','med_freq','spec_entr']
                 ]

# Trim to match actual feature count
all_feat_names = all_feat_names[:X.shape[1]]

loadings = pd.DataFrame(
    pca_full.components_[:2].T,
    columns=['PC1', 'PC2'],
    index=all_feat_names
)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, pc in zip(axes, ['PC1', 'PC2']):
    top = loadings[pc].abs().nlargest(n_top)
    vals = loadings.loc[top.index, pc]
    colors = ['#F44336' if v < 0 else '#2196F3' for v in vals]
    ax.barh(range(n_top), vals.values, color=colors, alpha=0.85)
    ax.set_yticks(range(n_top))
    ax.set_yticklabels(top.index, fontsize=7)
    ax.set_title(f'Top {n_top} Features for {pc}', fontweight='bold')
    ax.axvline(0, color='gray', lw=0.8)
    ax.set_xlabel('Loading')

plt.suptitle('PCA Feature Loadings', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
